# SLM gradient descent #

Current implementation of Fourier:
* N = 64: a couple of minutes
* N = 128: 40 minutes (scales like $N^4$)

### Plots of target and reconstruction. ###

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

target = np.loadtxt('slmOpt1_target.txt')
output = np.loadtxt('slmOpt1_output.txt')
plot_args = {'extent':[-1, 1, -1, 1],
             'interpolation':'None'}
```

### Absolute values of target and reconstruction: ###

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,12))
ax1.imshow(target, vmin=0, vmax=3, **plot_args)
ax1.set_title('Target')

ax2.imshow(output, vmin=0, vmax=3, **plot_args)
ax2.set_title('Reconstructed')
```

### Target and reconstruction plotted with autoscaling ###

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,12))
ax1.imshow(target, **plot_args)
ax1.set_title('Target')

ax2.imshow(output, **plot_args)
ax2.set_title('Reconstructed')
```

### Plot a cross-section through the middle: ###

```python
N = target.shape[0]
x = np.linspace(-1,1,N)
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(12,3))
# First plot:
idx = int(N/2)
ax1.plot(x, target[idx,:], '.-', label='Target')
ax1.plot(x, output[idx,:], '.-', label='Reconstruction')
ax1.legend()

# Second plot:
idx = int(N/3)
ax2.plot(x, target[idx,:], '.-', label='Target')
ax2.plot(x, output[idx,:], '.-', label='Reconstruction')
ax2.legend()

# Third plot:
idx = int(N/3)
ax3.plot(x, target[:,idx], '.-', label='Target')
ax3.plot(x, output[:,idx], '.-', label='Reconstruction')
ax3.legend()
```
